```python
from trafficgraphnn.genconfig import ConfigGenerator
from trafficgraphnn.sumo_network import SumoNetwork
```

```python
config = ConfigGenerator(net_name='smallnet_real')
```

```python
config.gen_grid_network(grid_number = 3, grid_length=600)
```

```python
config.gen_rand_trips(period=0.7, binomial=0, seed = 50, end_time = 3600, fringe_factor = 1000)
# from the sumo documentation:   
# --binomial=N          If this is set, the number of departures per seconds
#                         will be drawn from a binomial distribution with n=N
#                         and p=PERIOD/N where PERIOD is the argument given to
#                         option --period. Tnumber of attemps for finding a trip
#                         which meets the distance constraints
# This is done by calling the python file randomTrips.py which lives in the sumo/tools directory
```

```python
config.gen_e1_detectors(distance_to_tls=[5, 125], frequency=1)
config.gen_e2_detectors(distance_to_tls=0, frequency=1)
config.define_tls_output_file()
```

```python
sn = SumoNetwork.from_gen_config(config)
```

```python
sn.run()
```
