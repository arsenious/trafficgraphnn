```python
from trafficgraphnn.genconfig import ConfigGenerator
from trafficgraphnn.sumo_network import SumoNetwork
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from trafficgraphnn.liumethod import LiuEtAlRunner, LiuIntersection, LiuLane
from trafficgraphnn.preprocess_data import PreprocessData
import time

config = ConfigGenerator(net_name='simonnet')

# Parameters for network, trips and sensors (binomial must be an integer!!!)
end_time = 2000
config.gen_grid_network(grid_number = 3, grid_length=600, num_lanes=2, simplify_tls = True)
config.gen_rand_trips(period=0.2, binomial=2, seed = 50, end_time = end_time, fringe_factor = 1000)

config.gen_e1_detectors(distance_to_tls=[5, 125], frequency=1)
config.gen_e2_detectors(distance_to_tls=0, frequency=1)
config.define_tls_output_file()

# run the simulation to create output files
sn = SumoNetwork.from_gen_config(config, lanewise=True)
#sn.run()
### preprocess data ###
preprocess_handle = PreprocessData(sn)

preprocess_handle.preprocess_detector_data([], average_over_cycle=True)

```
